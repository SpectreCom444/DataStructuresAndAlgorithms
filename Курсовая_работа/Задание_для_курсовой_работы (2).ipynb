```python
def operation(n,a):
    if n == 1:#–∏—Å–∫–ª—é—á–µ–Ω–∏–µ –Ω–∞ —Å–ª—É—á–∞–π –≤–≤–æ–¥–∞ –æ–¥–Ω–æ–≥–æ —á–∏—Å–ª–∞
        print('1 1', -a[0], '1 1', '0', '1 1', '0', sep='\n')
        return

    print(1, n)#–≤—ã–±–∏—Ä–∞–µ–º –¥–∏–∞–ø–∞–∑–æ–Ω 
    for i in range(n):#–≤—ã–≤–æ–¥ —á–∏—Å–µ—Å–ª ùëèùëô,ùëèùëô+1,‚Ä¶,ùëèùëü –¥–ª—è –ø–µ—Ä–≤–æ–≥–æ –¥–µ–π—Å—Ç–≤–∏—è
        print(-a[i] * n, end = ' ')
        a[i] -= a[i] * n#–∏–∑–º–µ–Ω—è–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ —Ç–µ–∫—É—â–µ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ –º–∞—Å—Å–∏–≤–∞
    print()#–ø–µ—Ä–µ–≤–æ–¥–∏–º –Ω–∞ —Å–ª–µ–¥—É—é—â—É—é —Å—Ç—Ä–æ–∫—É

    print(1, n - 1)#–≤—ã–±–∏—Ä–∞–µ–º –¥–∏–∞–ø–∞–∑–æ–Ω
    for i in range(n - 1):#–≤—ã–≤–æ–¥ —á–∏—Å–µ—Å–ª ùëèùëô,ùëèùëô+1,‚Ä¶,ùëèùëü –¥–ª—è –≤—Ç–æ—Ä–æ–≥–æ –¥–µ–π—Å—Ç–≤–∏—è
        print(-a[i], end = ' ')
    print()#–ø–µ—Ä–µ–≤–æ–¥–∏–º –Ω–∞ —Å–ª–µ–¥—É—é—â—É—é —Å—Ç—Ä–æ–∫—É

    print(2, n)#–≤—ã–±–∏—Ä–∞–µ–º –¥–∏–∞–ø–∞–∑–æ–Ω
    for i in range(1, n):#–≤—ã–≤–æ–¥ —á–∏—Å–µ—Å–ª ùëèùëô,ùëèùëô+1,‚Ä¶,ùëèùëü –¥–ª—è —Ç—Ä–µ—Ç—å–µ–≥–æ –¥–µ–π—Å—Ç–≤–∏—è
        print(-a[i], end = ' ')
    print()#–ø–µ—Ä–µ–≤–æ–¥–∏–º –Ω–∞ —Å–ª–µ–¥—É—é—â—É—é —Å—Ç—Ä–æ–∫—É
    
#–í–≤–æ–¥ –¥–∞–Ω–Ω—ã—Ö
n = int(input())
a = list(map(int, input().split()))
#–í—ã–∑–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ operation –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –≤—ã—á–∏—Å–ª–µ–Ω–∏–π   
operation(n,a)
```

```python
def operation(n,l):
    #–æ–±—å—è–≤–ª–µ–Ω–∏–µ –ø–µ—Ä–µ–º–Ω–Ω—ã—Ö –º–µ—Ç–æ–¥–∞ 
    mx=0
    cur=0
    for i in range(0,n-1,2): #–∑–∞–ø—É—Å–∫–∞–µ–º —Ü–∏–∫–ª –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –ø–æ–¥–º–∞—Å—Å–∏–≤–∞ 1 —Å–ø–æ—Å–æ–±–æ–º
        cur+=l[i+1]-l[i]#–≤—ã—á–∏—Ç–∞–µ–º –∏–∑ –ø–æ—Å–ª–µ–¥—É—é—â–µ–≥–æ —Ç–µ–∫—É—â–µ–π —ç–ª–µ–º–µ–Ω—Ç –∏ –¥–æ–±–∞–≤–ª—è–µ–º —Ä–∞–∑–Ω—Å—Ç—å –∫ cur
        mx=max(cur,mx)#–∏—â–∏–º –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É—é –≤—ã–≥–æ–¥—É –æ—Ç —Ç–µ–∫—É—â–µ–≥–æ –≤–∞—Ä–∏–∞–Ω—Ç–∞ –ø–æ–¥–º–∞—Å—Å–∏–≤–∞
    cur=0
    for i in range(1,n-1,2):#–∑–∞–ø—É—Å–∫–∞–µ–º —Ü–∏–∫–ª –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –ø–æ–¥–º–∞—Å—Å–∏–≤–∞ 2 —Å–ø–æ—Å–æ–±–æ–º (–Ω–∞—á–∏–Ω–∞–µ–º —Å 1)
        cur+=l[i]-l[i+1]# –≤—ã—á–∏—Ç–∞–µ–º –∏–∑ —Ç–µ–∫—É—â–µ–≥–æ –ø–æ—Å–ª–µ–¥—É—é—â–∏–π —ç–ª–µ–º–µ–Ω—Ç –∏ –¥–æ–±–∞–≤–ª—è–µ–º —Å cur
        mx=max(cur,mx)#–∏—â–∏–º –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É—é –≤—ã–≥–æ–¥—É –æ—Ç —Ç–µ–∫—É—â–µ–≥–æ –≤–∞—Ä–∏–∞–Ω—Ç–∞ –ø–æ–¥–º–∞—Å—Å–∏–≤–∞
    s=0
    for i in range(0,n,2):#–ø—Ä–æ—Ö–æ–¥–∏–º –ø–æ –≤—Å–µ–º —á—ë—Ç–Ω—ã–º —ç–ª–µ–º–µ–Ω—Ç–∞–º
        s+=l[i]#—Å—É–º–º–∏—Ä—É–µ—Ç –≤—Å–µ —á—ë—Ç–Ω—ã–µ —ç–ª–µ–º–µ–Ω—Ç—ã
    return s+mx#–≤–æ–∑–≤—Ä–∞—â–∞–µ–º —Å—É–º–º—É —á—ë—Ç–Ω—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –∏ –Ω–∞–∏–±–æ–ª—å—à–µ–π –≤—ã–≥–æ–¥—ã –æ—Ç —Ä–∞–∑–≤—ë—Ä—Ç—ã–≤–∞–Ω–∏—è –ø–æ–¥–º–∞—Å—Å–∏–≤–∞

for _ in range(int(input())):#–≤—ã–∑—ã–≤–∞–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –Ω–∞–±–æ—Ä–∞ –≤–≤–µ–¥—ë–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
    #–í–≤–æ–¥ –¥–∞–Ω–Ω—ã—Ö
    n=int(input())
    l=list(map(int,input().split()))
    #–í—ã–∑–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ operation –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –≤—ã—á–∏—Å–ª–µ–Ω–∏–π   
    print(operation(n,l))
    
```

```python
def operation(n, k):
    a = []#–î–ª—è –∫–Ω–∏–≥ –∫–æ—Ç–æ—Ä—ã–µ –Ω—Ä–∞–≤—è—Ç—Å—è –ê–ª–∏—Å–µ
    b = []#–î–ª—è –∫–Ω–∏–≥ –∫–æ—Ç–æ—Ä—ã–µ –Ω—Ä–∞–≤—è—Ç—Å—è –ë–æ–±—É
    s = []
    for i in ' ' * n:
        t, x, y = map(int, input().split())
        if x & y:#–µ—Å–ª–∏ –∫–Ω–∏–≥–∞ –Ω—Ä–∞–≤–∏—Ç—Å—è –æ–±–æ–∏–º
            s += [t]
        elif x:#–ï—Å–ª–∏ –∫–Ω–∏–≥–∞ –Ω—Ä–∞–≤–∏—Ç—Å—è –ê–ª–∏—Å–µ
            a += [t]
        elif y:#–ï—Å–ª–∏ –∫–Ω–∏–≥–∞ –Ω—Ä–∞–≤–∏—Ç—Å—è –ë–æ–±—É
            b += [t]
            
    s += [i + j for i, j in zip(sorted(a), sorted(b))]# –¥–æ–±–∞–≤–ª—è–µ–º –≤ –º–∞—Å—Å–∏–≤ s –∫–Ω–∏–≥–∏ –∫–æ—Ç–æ—Ä—ã–µ –Ω—Ä–∞–≤—è—Ç—Å—è 1 –∏–ª–∏ 2 —Ä–µ–±—ë–Ω–∫—É
    return -1 if len(s) < k else sum(sorted(s)[:k])# –ø—Ä–æ–≤–µ—Ä—è–µ–º —á—Ç–æ –ø–æ–¥—Ö–æ–¥—è—â–µ–µ –º–Ω–æ–∂–µ—Å—Ç–≤–æ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –∏–ª–∏ –≤—ã–≤–æ–¥–∏–º -1
    
#–í–≤–æ–¥ –¥–∞–Ω–Ω—ã—Ö
n, k = map(int, input().split())
#–í—ã–∑–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ operation –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –≤—ã—á–∏—Å–ª–µ–Ω–∏–π
print(operation(n, k))
```

```python
from math import gcd #–ø–æ–¥–∫–ª—é—á–∞–µ–º –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è –ù–û–î

def operation(n, m, a, b):
    g = 0 #–ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è –¥–ª—è –ù–û–î –ø–µ—Ä–≤–æ–≥–æ –º–∞—Å—Å–∏–≤–∞
    for i in a[1:]:#—Ü–∏–∫–ª –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è –ù–û–î –ø–µ—Ä–≤–æ–≥–æ –º–∞—Å—Å–∏–≤–∞
        g = gcd(g,i-a[0])#–Ω–∞—Ö–æ–¥–∏–º –ù–û–î –ø–µ—Ä–≤–æ–≥–æ –º–∞—Å—Å–∏–≤–∞
    for i in range(m):#—Ü–∏–∫–ª –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è –ù–û–î –≤—Ç–æ—Ä–æ–≥–æ –º–∞—Å—Å–∏–≤–∞
        print(gcd(g,a[0]+b[i]),end = " ")#–Ω–∞—Ö–æ–¥–∏–º –ù–û–î –¥–ª—è an+bj –∏ –ù–û–î–∞ –ø–µ—Ä–≤–æ–≥–æ –º–∞—Å—Å–∏–≤–∞,–ø–æ–ª—É—á–∞—è —á–∏—Å–ª–∞ —Ä–∞–≤–Ω—ã–µ –ù–û–î(ùëé1+ùëèùëó,‚Ä¶,ùëéùëõ+ùëèùëó).

#–í–≤–æ–¥ –¥–∞–Ω–Ω—ã—Ö
n,m = map(int,input().split())
a = list(map(int,input().split()))
b = list(map(int,input().split()))
#–í—ã–∑–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ operation –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –≤—ã—á–∏—Å–ª–µ–Ω–∏–π
operation(n, m, a, b)
```

–í –ø–µ—Ä–≤–æ–º –ø—Ä–∏–º–µ—Ä–µ –º–æ–∂–Ω–æ –ø–µ—Ä–µ–≤–µ—Ä–Ω—É—Ç—å –ø–æ–¥–º–∞—Å—Å–∏–≤ [4,5]. –¢–æ–≥–¥–∞ ùëé=[2,3,2,3,1] –∏ 2‚ãÖ1+3‚ãÖ3+2‚ãÖ2+3‚ãÖ4+1‚ãÖ2=29.

–í–æ –≤—Ç–æ—Ä–æ–º –ø—Ä–∏–º–µ—Ä–µ –Ω–µ –Ω—É–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –æ–ø–µ—Ä–∞—Ü–∏—é. 13‚ãÖ2+37‚ãÖ4=174.

–í —Ç—Ä–µ—Ç—å–µ–º –ø—Ä–∏–º–µ—Ä–µ –º–æ–∂–Ω–æ –ø–µ—Ä–µ–≤–µ—Ä–Ω—É—Ç—å –ø–æ–¥–º–∞—Å—Å–∏–≤ [3,5]. –¢–æ–≥–¥–∞ ùëé=[1,8,3,6,7,6] –∏ 1‚ãÖ5+8‚ãÖ9+3‚ãÖ6+6‚ãÖ8+7‚ãÖ8+6‚ãÖ6=235.

```python
def operation(n,a,b):
    rez = 0 #–ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–π —Å—É–º–º—ã
    for i in range(n): 
        rez += a[i] * b[i] #–Ω–∞—Ö–æ–¥–∏–º –Ω–∞—á–∞–ª—å–Ω–æ–µ –∑–Ω–∞–µ—á–Ω–∏–µ –¥–ª—è –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π —Ä–µ–∑ –¥–æ —Ä–∞–∑–≤–µ—Ä—Ç—ã–≤–∞–Ω–∏—è –ø–æ–¥–º–∞—Å—Å–∏–≤–∞
    base = rez # —Å–æ–∑–¥–∞—ë–º –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –∫–æ—Ç–æ—Ä–∞—è —Å–æ—Ö—Ä–∞–Ω–∏—Ç –±–∞–∑–æ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –¥–ª—è –º–∞—Å—Å–∏–≤–∞ –¥–æ —Ä–∞–∑–≤—ë—Ä—Ç—ã–≤–∞–Ω–∏—è –ø–æ–¥–º–∞—Å—Å–∏–≤–∞
    for i in range(n): 
        for j in (i - 1, i): # —Ü–∏–∫–ª –∫–æ—Ç–æ—Ä—ã–π –ø–æ–∑–≤–æ–ª—è–µ—Ç –ø–æ–ª—É—á–∏—Ç—å –≤–∞—Ä–∏–∞–Ω—Ç—ã –ø–æ–¥–º–∞—Å—Å–∏–≤–æ–≤ –Ω–∞—á–∏–Ω–∞—é—â–∏—Ö—Å—è —Å-
            # - –Ω–µ—á—ë—Ç–Ω–æ–≥–æ/—á—ë—Ç—Ç–Ω–æ–≥–æ –∏–Ω–¥–µ–∫—Å–∞ –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è –∏ –ø—Ä–æ–≤–µ—Ä–∫–∏ –≤—Å–µ—Ö –≤–∞—Ä–∏–∞–Ω—Ç–æ–≤ –ø–æ–¥–º–∞—Å—Å–∏–≤–∞
            cur = base #–ø—Ä–∏—Å–≤–∞–∏–≤–∞–µ–º –Ω–æ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π –∑–Ω–∞—á–µ–Ω–∏–µ rez —á—Ç–æ-–±—ã –±—ã–ª–∞ –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å –ø–æ–ª—É—á–∏—Ç—å –∑–Ω–∞–µ—á–Ω–∏–µ –ø–æ—Å–ª–µ –∏–∑–º–µ–Ω–µ–Ω–∏–π –æ–±—Ä–∞—Ç–Ω–æ
            k = i + 1 
            while j >= 0 and k < n: # –∑–∞–ø—É—Å–∫–∞–µ–º —Ü–∏–∫–ª –∫–æ—Ç–æ—Ä—ã–π –ø—Ä–æ–≤–µ—Ä–∏—Ç –≤—Å–µ –≤–∞—Ä–∏–∞–Ω—Ç—ã –ø–æ–¥–º–∞—Å—Å–∏–≤–∞ —Å —Ç–µ–∫—É—â–µ–≥–æ –∏–Ω–¥–µ–∫—Å–∞
                cur -= (a[j] - a[k]) * (b[j] - b[k]) # —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º –≤—ã–≥–æ–¥—É –æ—Ç —Ä–∞—Å—à–∏—Ä–µ–Ω–∏—è –ø–æ–¥–º–∞—Å—Å–∏–≤–∞ –Ω–∞ 2 —Ç–µ–∫—É—â–∏—Ö –∑–Ω–∞—á–µ–Ω–∏—è
                rez = max(rez, cur) # –∏–∑–º–µ–Ω—è–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∏—Ä—É—é—â–∏–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π –µ—Å–ª–∏ cur –ë–æ–ª—å—à–µ –µ—ë —Ç–µ–∫—É—â–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è
                j -= 1 # —Ä–∞—Å—à–∏—Ä—è–µ–º –ø–æ–¥–º–∞—Å—Å–∏–≤ –Ω–∞ 1 –≤–ª–µ–≤–æ
                k += 1 # —Ä–∞—Å—à–∏—Ä—è–µ–º –ø–æ–¥–º–∞—Å—Å–∏–≤ –Ω–∞ 1 –≤–ø—Ä–∞–≤–æ
    return rez #–≤–æ–∑–≤—Ä–∞—â–∞–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç

#–í–≤–æ–¥ –¥–∞–Ω–Ω—ã—Ö
n = int(input())
a = list(map(int, input().split()))
b = list(map(int, input().split()))
#–í—ã–∑–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ operation –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –≤—ã—á–∏—Å–ª–µ–Ω–∏–π
print(operation(n,a,b))
```

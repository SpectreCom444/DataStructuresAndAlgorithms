### Задание №2
Написать генератор случайных матриц(многомерных), который принимает
опциональные параметры <b>m</b>, <b>n</b>, <b>min_limit</b>, <b>max_limit</b>, где <b>m</b> и <b>n</b> указывают размер
матрицы, а <b>min_lim</b> и <b>max_lim</b> - минимальное и максимальное значение для
генерируемого числа.

```python
import random
import time
user_m,user_n,user_min_limit,user_max_limit = int(input('n')),int(input('m')),int(input('min')),int(input('max'))
def generator(user_m,user_n,user_min_limit,user_max_limit):
    return [ [ random.randint(user_min_limit, user_max_limit) for j in range(user_n)] for i in range(user_m) ]
print('Matrix:',generator(user_m,user_n,user_min_limit,user_max_limit))
```

### Задание №3
Реализовать методы сортировки строк числовой матрицы в соответствии с
заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со
временем стандартной функции сортировки. Испытания проводить на сгенерированных
матрицах.

```python
import copy
```

```python
def matrix_gen(n,min_limit,max_limit):
    matrix = generator(n,1,min_limit,max_limit)
    mas = [matrix[i][0] for i in range(n) ]
    return mas
```

```python
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
start_time = time.time()
mas=matrix_gen(200,-250,1000)
print('Source array:\n',*mas)
def vibor(mas):
    for i in range(len(mas) - 1):
        m = i
        j = i + 1
        while j < len(mas):
            if mas[j] < mas[m]:
                m = j
            j = j + 1
        mas[i], mas[m] = mas[m], mas[i]
    return mas
print ('Sorted matrix:\n',*vibor(mas))
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка вставкой.
start_time = time.time()
mas=matrix_gen(200,-250,1000)
print('Source array:\n',*mas)
def vst(mas):
    for i in range(1, len(mas)): 
        value = mas[i] 
        j = i - 1 
        while j >= 0 and value < mas[j]: 
            mas[j + 1] = mas[j] 
            j -= 1 
        mas[j + 1] = value 
    return mas
print ('Sorted matrix:\n',*vst(mas))
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка обменом. || Сортировка пузырьком.
start_time = time.time()
mas=matrix_gen(200,-250,1000)
print('Source array:\n',*mas)
def obmen(mas):
    swap=True
    while swap:
        swap = False
        for i in range(len(mas) - 1):
            if mas[i] > mas[i + 1]:
                mas[i], mas[i + 1] = mas[i + 1], mas[i]
                swap = True
    return mas
print ('Sorted matrix:\n',*obmen(mas))
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка Шелла.
start_time = time.time()
mas=matrix_gen(200,-250,1000)
print('Source array:\n',*mas)
def shel(mas):
    inc = len(mas) // 2
    while inc:
        for i, n in enumerate(mas):
            while i >= inc and mas[i - inc] > n:
                mas[i] = mas[i - inc]
                i -= inc
            mas[i] = n
        inc = 1 if inc == 2 else int(inc * 5.0 / 11)
    return mas
print ('Sorted matrix:\n',*shel(mas))
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Быстрая сортировка.
start_time = time.time()
mas=matrix_gen(200,-250,1000)
print('Source array:\n',*mas)
def quicksort(mas):
    if len(mas) <= 1:
        return mas
    else:
        q = random.choice(mas)
    a_mas = [n for n in mas if n < q]
 
    r_mas = [q] * mas.count(q)
    b_mas = [n for n in mas if n > q]
    return quicksort(a_mas) + r_mas + quicksort(b_mas)
print ('Sorted matrix:\n',*quicksort(mas))
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
    
```

```python
# Турнирная сортировка.
import heapq
start_time = time.time()
mas=matrix_gen(200,-200,1000)
print('Source array:\n',*mas)
def turnir(mas):
    size=len(mas)
    max_s=len(mas)+1
    h=[]
    while len(h)<max_s and len(mas)>0:
        heapq.heappush(h,mas.pop(0))
    win=[]
    los=[]
    while len(mas)>0:
        if len(win)==0:
            win.append(heapq.heappop(h))
        if mas[0]>win[-1]:
            heapq.heappush(h,mas.pop(0))
        else:
            los.append(mas.pop(0))
                    
            if len(h)>0:
                win.append(heapq.heappop(h))
    while len(h)>0:
        win.append(heapq.heappop(h))
    if len(los)==0:
        return win
    los=turnir(los)
    def merge(mas_a,mas_b,res):
        a,b,r=0,0,0
        while a<len(mas_a) and b<len(mas_b):
            if mas_a[a]<mas_b[b]:
                res[r]=mas_a[a]
                a+=1
            else:
                res[r]==mas_b[b]
                b+=1
            r+=1
        while a<len(mas_a):
            res[r]==mas_a[a]
            a+=1
            r+=1
        while b<len(mas_b):
            res[r]=mas_b[b]
            b+=1
            r+=1
                
    mas=list(range(size))
    merge(los,win,mas)
    return mas  
    
print ('Sorted matrix:\n',*turnir(mas))
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

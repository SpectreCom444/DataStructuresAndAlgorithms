```python
import random
n=20
min_limit = -250
max_limit = 1011
list_n=[ random.randint(min_limit, max_limit) for j in range(n)]
print(list_n)
print (vibor(list_n))
print (vst(list_n))
print (obmen(list_n))
print (shel(list_n))
print (quicksort(list_n))

def vibor(list_n):
    list_r=list_n
    for i in range(len(list_r) - 1):
        m = i
        j = i + 1
        while j < len(list_r):
            if list_r[j] < list_r[m]:
                m = j
            j = j + 1
        list_r[i], list_r[m] = list_r[m], list_r[i]
    return list_r
def vst(list_n):
    list_r=list_n
    for i in range(1, len(list_r)): 
        value = list_r[i] 
        j = i - 1 
        while j >= 0 and value < list_r[j]: 
            list_r[j + 1] = list_r[j] 
            j -= 1 
        list_r[j + 1] = value 
    return list_r
def obmen(list_n):
    list_r=list_n
    swap=True
    while swap:
        swap = False
        for i in range(len(list_r) - 1):
            if list_r[i] > list_r[i + 1]:
                list_r[i], list_r[i + 1] = list_r[i + 1], list_r[i]
                swap = True
    return list_r
def shel(list_n):
    list_r=list_n
    value = len(list_r) // 2
    while value:
        for i, el in enumerate(list_r):
            while i >= value and list_r[i - value] > el:
                list_r[i] = list_r[i - value]
                i -= value
            list_r[i] = el
        value = 1 if value == 2 else int(value * 5.0 / 11)
    return list_r
def quicksort(nums):
    if len(nums) <= 1:
        return nums
    else:
        q = random.choice(nums)
        a_nums = []
        b_nums = []
        c_nums = []
        for n in nums:
            if n < q:
                a_nums.append(n)
            elif n > q:
                b_nums.append(n)
            else:
                c_nums.append(n)
        return quicksort(a_nums) + c_nums + quicksort(b_nums)
```
